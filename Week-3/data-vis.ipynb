```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.figure_factory as ff
```

```python
birds_df = pd.read_csv("../data/birds.csv")
```

```python
birds_df.head()
```

```python
plt.pie(birds_df['MinBodyMass'], labels=birds_df['Genus'])
plt.title('Min Body Mass Distribution by ')
plt.show()
```

```python
px.treemap(birds_df,path=[px.Constant("Birds"), 'Family', 'Genus'])
```

```python
plt.figure(figsize=(15,6))
plt.plot(birds_df['MaxWingspan'], birds_df['Genus'], color="red")
```

```python
sns.countplot(x='Family', 
              data=birds_df
              )
```

```python
plt.stem( 'Family', 
         'MinLenght',  
         linefmt='--',  
        data=birds_df
        )
plt.show()
```

```python
sns.set(style="darkgrid")

sns.boxplot(x=birds_df['ConservationStatus'], 

            y=birds_df['MinBodyMass'],

            hue=birds_df['ConservationStatus'],

            palette='rocket')
```

```python
sns.histplot(data=birds_df, x='MinLength', kde=True)
```

```python
plt.bar(birds_df['Category'], birds_df['MinLength'], color='skyblue')
plt.xlabel('Category')
plt.ylabel('Minlength')
plt.title('Categor - Min Length')
plt.show()
```
