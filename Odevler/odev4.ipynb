```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
```

```python
url = "https://hp-api.onrender.com/api/characters"
```

```python
try:
    # Send a GET request
    response = requests.get(url)
    
    if response.status_code == 200:
        characters_data = response.json()
        print("Harry Potter Characters:")
        for character in characters_data:
            print(f"Name: {character['name']}")
    else:
        print(f" Hata kodu : {response.status_code}")
except Exception as e:
    print(f"Hata : {str(e)}")
```

```python
df = pd.DataFrame(characters_data)
df
```

```python
df["species"].value_counts()
```

```python
# Plot a pie chart for the 'species' column
human_count = df[df['species'] == 'human'].shape[0]
labels = ['Human', 'Others']
sizes = [human_count, len(df) - human_count]
explode = (0.1, 0)  # To slightly separate the slices

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Species Distribution')
plt.show()
```

```python
# Replace empty strings with NaN in the 'house' column
df['house'].replace('', "Unknown", inplace=True)

# Plot a bar plot for the 'house' column
plt.figure(figsize=(10, 6))
sns.countplot(x='house', data=df, palette='viridis', order=df['house'].value_counts().index)
plt.title('House Distribution')
plt.xlabel('House')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
gender_counts = df['gender'].value_counts()
gender_counts
```

```python
plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.title('Gender Distribution')
plt.show()
```

```python
 # Plot a swarm plot for the 'wizard' column
plt.figure(figsize=(10, 6))
sns.swarmplot(x='wizard', data=df, palette='Set2')
plt.title('Wizard Distribution')
plt.xlabel('Wizard')
plt.ylabel('Some Numeric Column')
plt.show()
```

```python
""" 
Parameters:
    - data: The dataset you want to visualize.
    - x: Specifies the data on the x-axis.
    - bins: Determines the number of bins for the histogram bars. A higher value provides more detail.
    - kde: Shows the Kernel Density Estimate (KDE) plot. This represents the data distribution more smoothly.
    - color: Specifies the color of the histogram bars.
    - multiple: If set to "stack", it stacks histograms of different categories.
    - element: Specifies the type of histogram element. It can take values like "step", "poly", "stepfilled," etc.
    - stat: Determines what the bar heights represent, such as "count", "density", or "probability".
    - common_norm: If True, normalizes the sum of different subgroups to 1.

"""

# Create a distribution plot for 'yearOfBirth'
plt.figure(figsize=(12, 6))
ax = sns.histplot(df['yearOfBirth'].dropna(), kde=True, color='skyblue', bins=20)
plt.title('Distribution of Year of Birth')
plt.xlabel('Year of Birth')
plt.ylabel('Frequency')

# Add text to the plot to display the total number of values
total_values = df['yearOfBirth'].count()
ax.text(0.98, 0.9, f"Total: {total_values}", transform=ax.transAxes, ha='right', fontsize=10)


plt.show()
```
